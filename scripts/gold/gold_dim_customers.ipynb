#The Transformation Logic

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id,
    ci.customer_number,
    ci.first_name,
    ci.last_name,
    la.country,
    ci.marital_status,
    CASE
        WHEN ci.gender <> 'n/a' THEN ci.gender
        ELSE COALESCE(ca.gender, 'n/a')
    END AS gender,
    ca.birth_date AS birthdate,
    ci.created_date AS create_date
FROM erp.silver.crm_customers ci
LEFT JOIN erp.silver.erp_customers ca
    ON ci.customer_number = ca.customer_number
LEFT JOIN erp.silver.erp_customer_location la
    ON ci.customer_number = la.customer_number
"""
df = spark.sql(query)

     

df.limit(10).display()
     
```

#Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("erp.gold.dim_customers")

#Sanity checks of Gold table
df.limit(10).display()
```
