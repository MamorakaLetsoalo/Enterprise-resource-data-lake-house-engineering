#The Transformation Logic

```python
query = """
SELECT
    sd.order_number,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM erp.silver.crm_sales sd
LEFT JOIN erp.gold.dim_products pr
    ON sd.product_number = pr.product_number
LEFT JOIN erp.gold.dim_customers cu
    ON sd.customer_id = cu.customer_id;
"""
df = spark.sql(query)

df.limit(10).display()
```

#Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("erp.gold.fact_sales")
     
#Sanity checks of Gold table
df.limit(10).display()
```
